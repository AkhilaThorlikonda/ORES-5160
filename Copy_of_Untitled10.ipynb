```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/content/covid2.csv")
df
```

```python
df = df.drop(columns=["New cases","New deaths","New recovered"])
df
```

```python
df[["Confirmed", "Deaths", "Recovered", "Active"]] = df[["Confirmed", "Deaths", "Recovered", "Active"]].apply(pd.to_numeric)
df
```

```python
df ["Deaths"] = df ["Deaths"].replace(0,'')
df ["Recovered"] = df ["Recovered"].replace(0,'')
df ["Active"] = df ["Active"].replace(0,'')
df ["Deaths / 100 Cases"] = df ["Deaths / 100 Cases"].replace(0.00,'')
df ["Deaths / 100 Recovered"] = df ["Deaths / 100 Recovered"].replace(0.00,'')
df ["1 week change"] = df ["1 week change"].replace(0,'')
df ["1 week % increase"] = df ["1 week % increase"].replace(0,'')
df
```

```python
df = df.rename(columns={"Deaths / 100 Cases": "Death Rate (%)", "Recovered / 100 Cases": "Recovery Rate (%)"})
df
```

```python
df['Deaths'] = pd.to_numeric(df['Deaths'], errors='coerce')
df['Confirmed'] = pd.to_numeric(df['Confirmed'], errors='coerce')
df.dropna(subset=['Deaths', 'Confirmed'], inplace=True)
df['Mortality Rate'] = (df['Deaths'] / df['Confirmed']) * 100
print(df)
```

```python
df = df.drop_duplicates()
df
```

```python
Q1 = df["Confirmed"].quantile(0.25)
Q3 = df["Confirmed"].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = df[(df["Confirmed"] < lower_bound) | (df["Confirmed"] > upper_bound)]
print("Potential Outliers in the 'Confirmed' Column:")
print(outliers)
df
```

```python
df.columns = df.columns.str.lower().str.replace(' ', '_')
df
```
